```python
from sklearn.model_selection import train_test_split
```

```python
img_list = glob(imgs_dir + '/*')
train_img, valid_img = train_test_split(img_list, test_size = 0.1, random_state = 42)
train_img, test_img = train_test_split(train_img, test_size = 0.2, random_state = 42)
```

```python
with open('/dataset/train.txt', 'w') as f:
    f.write('\n'.join(train_img) + '\n')
with open('/dataset/val.txt', 'w') as f:
    f.write('\n'.join(valid_img) + '\n')
with open('/dataset/test.txt', 'w') as f:
    f.write('\n'.join(test_img) + '\n')
```

```python
import yaml
```

```python
with open('/dataset/data.yaml', 'w') as f:
    data = {
        'train' : '/dataset/train.txt',
        'val' : '/dataset/val.txt',
        'test': '/dataset/test.txt',
        'nc' : 2,
        'names' : ['cat', 'dog']
    }
    yaml.dump(data, f)
```
